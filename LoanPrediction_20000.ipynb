```python
df1 = df.drop(['name','identification_through'],axis = 1)
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
X_train,X_test,y_train,y_test = train_test_split(df1.drop('loan_default',axis= 1),df1['loan_default'],test_size = 0.3,random_state=42)
```

```python
cat_cols = X_train.select_dtypes(include=['object']).columns
num_cols = X_train.select_dtypes(exclude=['object']).columns
```

```python
preprocessor = ColumnTransformer([
    
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
    ('num', 'passthrough', num_cols)

])
```

```python
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', DecisionTreeClassifier(max_depth=5, random_state=42))
])

pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
with open('loan_prediction_pipeline_20000.pkl', 'wb') as f:
    pickle.dump(pipe, f)
```
